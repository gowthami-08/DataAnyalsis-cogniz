```python
import pandas as pd


df = pd.read_csv("Dataset .csv")


print(df.columns)
```

```python

city_col = 'City'
rating_col = 'Aggregate rating'

df = df.dropna(subset=[city_col, rating_col])
```

```python
top_city_counts = df[city_col].value_counts()
most_restaurants_city = top_city_counts.idxmax()
print(f"City with most restaurants: {most_restaurants_city} ({top_city_counts.max()} restaurants)")
```

```python

city_avg_rating = df.groupby(city_col)[rating_col].mean().sort_values(ascending=False)


print("Top Cities by Average Rating:")
print(city_avg_rating.head(5))
```

```python
highest_rated_city = city_avg_rating.idxmax()
highest_rating = city_avg_rating.max()

print(f"City with highest average rating: {highest_rated_city} ({round(highest_rating, 2)})")
```

```python
import matplotlib.pyplot as plt


top_city_counts.head(10).plot(kind='bar', color='orange', figsize=(8,5))
plt.title("Top 10 Cities with Most Restaurants")
plt.ylabel("Number of Restaurants")
plt.xlabel("City")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()


city_avg_rating.head(10).plot(kind='bar', color='green', figsize=(8,5))
plt.title("Top 10 Cities by Average Rating")
plt.ylabel("Average Rating")
plt.xlabel("City")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
